```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.merge(study_results, mouse_metadata, how='left', on='Mouse ID')

# Display the data table for preview
mouse_df.head()
```

```python
# Checking the number of mice.
unique_mice = len(mouse_df['Mouse ID'].unique())

unique_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_duplicate = mouse_df.loc[mouse_df.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()

mouse_duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_duplicate_df = mouse_df[mouse_df["Mouse ID"] == "g989"] 

mouse_duplicate_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_cleaned_df = mouse_df[mouse_df["Mouse ID"] != "g989"]

mouse_cleaned_df
```

```python
# Checking the number of mice in the clean DataFrame.

unique_mice_cleaned = len(mouse_cleaned_df['Mouse ID'].unique())

unique_mice_cleaned
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.

summary_df = mouse_cleaned_df

sum_mean = summary_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
sum_med = summary_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
sum_var = summary_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
sum_std = summary_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
sum_sem = summary_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

#create summary table
summary_df = pd.DataFrame({
    "Mean" : sum_mean,
    "Median" : sum_med,
    "Variance" : sum_var, 
    "St Dev" : sum_std, 
    "SEM" : sum_sem
})

summary_df = summary_df.round(2)

summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse

max_timepoint = mouse_cleaned_df.groupby("Mouse ID").max()
max_timepoint = max_timepoint.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

merge_df = max_timepoint[["Mouse ID","Timepoint"]].merge(mouse_cleaned_df, on=["Mouse ID","Timepoint"], how='left')

merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Capomulin, Ramicane, Infubinol, and Ceftamin

treatments_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)

tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_drug = merge_df['Tumor Volume (mm3)'].loc[merge_df['Drug Regimen'] == drug]    
    
    # add subset 
    tumor_vol_data.append(tumor_vol_drug)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_vol_drug.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f"For {drug} IQR is {iqr}")
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"For {drug} values below {lower_bound} could be outliers")
    print(f"For {drug} values above {upper_bound} could be outliers\n")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
flierprops = dict(marker='o', markerfacecolor='green', markersize=10, linestyle='none')

fig, ax = plt.subplots()
ax.set_title("Final Tumor Tolume per Regimens")
ax.set_ylabel("Tumor Volume (mm3)")
ax.boxplot(tumor_vol_data, flierprops=flierprops)
ax.set_xticklabels(treatments_list)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

mouse_r944 = mouse_cleaned_df.loc[mouse_cleaned_df['Mouse ID'] == 'r944']

plt.plot(mouse_r944["Timepoint"], mouse_r944["Tumor Volume (mm3)"])
plt.title("Timepoint versus Tumor Volume (mm3) of mouse r944")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

capomulin_df = mouse_cleaned_df.loc[mouse_cleaned_df['Drug Regimen'] == 'Capomulin']

capomulin_avg = capomulin_df.groupby(["Weight (g)"]).mean()
capomulin_avg = capomulin_avg.reset_index()

plt.scatter(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"] )
plt.title("Weight vs Tumor Volume for Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = capomulin_avg["Weight (g)"]
y_values = capomulin_avg["Tumor Volume (mm3)"]
(slope, intercept,rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq= f"y = {round(slope, 2)} x + {round(intercept, 2)}"

plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, color='red')
plt.annotate(line_eq, (20.5,36.5), fontsize=15, color='red')
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Weight vs Tumor Volume for Capomulin")
plt.show()

print(f"Correlation coefficient is {round(rvalue,2)}")
```
